```python
import pandas as pd
from datetime import datetime
```

```python
# import data
videos = pd.read_csv('Table data.csv')
```

```python
videos.head(5)
```

```python
# remove first row
videos = videos.drop([0])
```

```python
videos = videos.rename(columns={"Impressions click-through rate (%)":"ctr","Estimated revenue (GBP)":"revenue","Video publish time":"published"})
```

```python
# Calculate the 90th percentile of views
percentile_90 = videos['ctr'].quantile(0.9)
```

```python
# Calculate the 90th percentile of views
percentile_90_views = videos['Views'].quantile(0.9)
```

```python
# Calculate top earners
earners_90 = videos['revenue'].quantile(0.9)
```

```python
# Add a new column for the top earners
videos['top_earners'] = videos['revenue'].apply(lambda x: True if x >= earners_90 else False)
```

```python
# Add age
# Convert the 'published_date' column to datetime
videos['published'] = pd.to_datetime(videos['published'])
```

```python
# Calculate the age of the video as the difference between now and the publication date
current_date = datetime.now()
videos['age'] = current_date - videos['published']
```

```python
# Convert age to a more readable format (like number of days)
videos['age_days'] = videos['age'].dt.days

# Display the DataFrame
print(videos)
```

```python
top_earners = videos[videos['top_earners'] == True]
len(top_earners)
```

```python
# Add a new column for the top 10% flag
videos['best_ctr'] = videos['ctr'].apply(lambda x: True if x >= percentile_90 else False)
```

```python
best_ctr = videos[videos['best_ctr'] == True]
len(best_ctr)
```

```python
# Add a new column for the top 10% most popular (highest views)
videos['popular'] = videos['Views'].apply(lambda x: True if x >= percentile_90_views else False)
```

```python
popular = videos[videos['popular'] == True]
len(popular)
```
